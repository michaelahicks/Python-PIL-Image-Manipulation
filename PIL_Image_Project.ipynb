```python
import PIL
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageDraw, ImageFont, ImageOps

# read image and convert to RGB
file = "readonly/msi_recruitment.gif"
img = Image.open(file).convert('RGB')

# build a list of 9 images which have different brightnesses
#enhancer=ImageEnhance.Color(image)
images=[]
text_strings = []
rgba_vals = []
border = (0, 0, 0, 50)
img = ImageOps.expand(img, border = border)

width = img.size[0]
height = img.size[1]
text_box = (800, 50)
#print(width, height)

r1_img = img.copy()
r2_img = img.copy()
r3_img = img.copy()
g1_img = img.copy()
g2_img = img.copy()
g3_img = img.copy()
b1_img = img.copy()
b2_img = img.copy()
b3_img = img.copy()

for x in range(3):
    if x == 0:
        channel = 0
    elif x == 1:
        channel = 1
    else:
        channel = 2
    #print(channel)
    for y in range(3):
        #print("*"*10)
        #print(y)
        if y == 0:
            intensity = 0.1
        elif y == 1:
            intensity = 0.5
        else:
            intensity = 0.9        
        outstring = ("channel {} intensity {}".format(channel, intensity))
        text_strings.append(outstring)

#for i in range(1, 10):
#    image.ImageDraw.text((0,0), "channel {} intensity {}", fill = (200, 100, 50))
#    images.append(enhancer.color(i/10))

for i in range(0, width):
    for j in range(0, height):
        data = r1_img.getpixel((i, j))
        r1 = data[0]*0.1
        r1 = int(r1)
        r1_img.putpixel((i,j), (r1, data[1], data[2]))
images.append(r1_img)
r1_txt = int(255*0.1)
rgba_vals.append(r1_txt)

for i in range(0, width):
    for j in range(0, height):
        data = r2_img.getpixel((i, j))
        r2 = data[0]*0.5
        r2 = int(r2)
        r2_img.putpixel((i,j), (r2, data[1], data[2]))
images.append(r2_img)
r2_txt = int(255*0.5)
rgba_vals.append(r2_txt)

for i in range(0, width):
    for j in range(0, height):
        data = r3_img.getpixel((i, j))
        r3 = data[0]*0.9
        r3 = int(r3)
        r3_img.putpixel((i,j), (r3, data[1], data[2]))
images.append(r3_img)
r3_txt = int(255*0.9)
rgba_vals.append(r3_txt)

for i in range(0, width):
    for j in range(0, height):
        data = g1_img.getpixel((i, j))
        g1 = data[1]*0.1
        g1 = int(g1)
        g1_img.putpixel((i,j), (data[0], g1, data[2]))
images.append(g1_img)
g1_txt = int(255*0.1)
rgba_vals.append(g1_txt)

for i in range(0, width):
    for j in range(0, height):
        data = g2_img.getpixel((i, j))
        g2 = data[1]*0.5
        g2 = int(g2)
        g2_img.putpixel((i,j), (data[0], g2, data[2]))
images.append(g2_img)
g2_txt = int(255*0.5)
rgba_vals.append(g2_txt)

for i in range(0, width):
    for j in range(0, height):
        data = g3_img.getpixel((i, j))
        g3 = data[1]*0.9
        g3 = int(g3)
        g3_img.putpixel((i,j), (data[0], g3, data[2]))
images.append(g3_img)
g3_txt = int(255*0.9)
rgba_vals.append(g3_txt)

for i in range(0, width):
    for j in range(0, height):
        data = b1_img.getpixel((i, j))
        b1 = data[2]*0.1
        b1 = int(b1)
        b1_img.putpixel((i,j), (data[0], data[1], b1))
images.append(b1_img)
b1_txt = int(255*0.1)
rgba_vals.append(b1_txt)

for i in range(0, width):
    for j in range(0, height):
        data = b2_img.getpixel((i, j))
        b2 = data[2]*0.5
        b2 = int(b2)
        b2_img.putpixel((i,j), (data[0], data[1], b2))
images.append(b2_img)
b2_txt = int(255*0.5)
rgba_vals.append(b2_txt)

for i in range(0, width):
    for j in range(0, height):
        data = b3_img.getpixel((i, j))
        b3 = data[2]*0.9
        b3 = int(b3)
        b3_img.putpixel((i,j), (data[0], data[1], b3))
images.append(b3_img)
b3_txt = int(255*0.9)
rgba_vals.append(b3_txt)

for x in range(9):
    with images[x].convert("RGBA") as base:
        txt = Image.new('RGBA', img.size, (225, 255, 255, 1))
        fnt =ImageFont.truetype("readonly/fanwood-webfont.ttf", 50)
        d = ImageDraw.Draw(txt)
        if x < 3:
            d.text((1,455), text_strings[x], font = fnt, fill = (rgba_vals[x], 255, 255, 255))
        elif x > 2 and x < 6:
            d.text((1,455), text_strings[x], font = fnt, fill = (255, rgba_vals[x], 255, 255))
        else:
            d.text((1,455), text_strings[x], font = fnt, fill = (255, 255, rgba_vals[x], 255))
        images[x] = Image.alpha_composite(base, txt)
        #display(out)
        
# create a contact sheet from different brightnesses
first_image=images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))
x=0
y=0

for img in images:
    # Lets paste the current image into the contact sheet
    contact_sheet.paste(img, (x, y) )
    # Now we update our X position. If it is going to be the width of the image, then we set it to 0
    # and update Y as well to point to the next "line" of the contact sheet.
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height
    else:
        x=x+first_image.width

# resize and display the contact sheet
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
